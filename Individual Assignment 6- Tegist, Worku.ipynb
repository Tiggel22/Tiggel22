# Individual Assignment 6: BMI Calculators

This program calculates the BMI for individuals listed in a .csv. It will display the statistics of weights, heights, & BMI for female, male, and combined.

- Input: .csv file 
- Output: Statistics for male & female dataset
- Author: Tegist Worku
- Date: 26 June 2023

```python
# import necessary libraries 
import numpy as np
import pandas as pd

# function to calculate BMI
def calculateBMI(height, weightlbs):
    weightKg = weightlbs/2.2046
    heightM2 = np.power(height, 2)
    BMI = weightKg/heightM2
    return BMI

# read in female dataset, add a 'Gender' column, set 'Gender' as index
femaleData = pd.read_csv('measurements-female.csv')
femaleData['Gender'] = 'Female'
femaleData = femaleData.set_index('Gender')

# convert centimeters to meters in height column & drop centimeter column
femaleData['Height (m)'] = np.array(femaleData['Height (cm)'])*0.01
femaleData = femaleData.drop(['Height (cm)'], axis=1)

# read in male dataset, add a 'Gender' column, set 'Gender' as index
maleData = pd.read_csv('measurements-male.csv')
maleData['Gender'] = 'Male'
maleData = maleData.set_index('Gender')

# repeat the same centimeters to meter conversion
maleData['Height (m)'] = np.array(maleData['Height (cm)'])*0.01
maleData = maleData.drop(['Height (cm)'], axis=1)

# use function calculate BMI for male & female
femaleData['BMI'] = calculateBMI(femaleData['Height (m)'], femaleData['Weight (lb)'])
print(f"The BMI for each individual in the female data set are as follows: \n{femaleData}")

maleData['BMI'] = calculateBMI(maleData['Height (m)'], maleData['Weight (lb)'])
print(f"The BMI for each individual in the male data set are as follows: \n{maleData}")

# use aggregation to show the stats for each column
print("STATISTICS FOR FEMALE DATASETS")
print("The mean for BMI, height(m), weight(lbs): \n"+str(femaleData.mean(0)))
print("The median for BMI, heights(m), weights(lbs): \n"+str(femaleData.median(0)))
print("The maximum for BMI, heights(m), weights(lbs):\n"+str(femaleData.max(0)))
print("The minimum for BMI, heights(m), weights(lbs):\n"+str(femaleData.min(0)))
print("The standard deviation for BMI, heights(m), weights(lbs):\n"+str(femaleData.std(0)))
print("\n")
print("STATISTICS FOR MALE DATASETS")
print("The mean for BMI, height(m), weight(lbs): \n"+str(maleData.mean(0)))
print("The median for BMI, heights(m), weights(lbs): \n"+str(maleData.median(0)))
print("The maximum for BMI, heights(m), weights(lbs):\n"+str(maleData.max(0)))
print("The minimum for BMI, heights(m), weights(lbs):\n"+str(maleData.min(0)))
print("The standard deviation for BMI, heights(m), weights(lbs):\n"+str(maleData.std(0)))
print("\n")
# combine the two datasets and create a new dataframe
combinedData = pd.concat([femaleData, maleData])
print("STATISTICS FOR COMBINED MALE & FEMALE DATASETS")
print("The mean for BMI, height(m), weight(lbs): \n"+str(combinedData.mean(0)))
print("The median for BMI, heights(m), weights(lbs): \n"+str(combinedData.median(0)))
print("The maximum for BMI, heights(m), weights(lbs):\n"+str(combinedData.max(0)))
print("The minimum for BMI, heights(m), weights(lbs):\n"+str(combinedData.min(0)))
print("The standard deviation for BMI, heights(m), weights(lbs):\n"+str(combinedData.std(0)))
print("\n")

# get the total number of male & female that are underweight
underweightFemale = np.sum(18.5>femaleData['BMI'])
underweightMale = np.sum(18.5>maleData['BMI'])
totalUnderweight =underweightFemale + underweightMale 

print(f"{underweightFemale} female individuals belong to the 'Underweight' category.")
print(f"{underweightMale} male individuals belong to the 'Underweight' category.")
print(f"With a total of {totalUnderweight} male & female individuals that are classified as 'Underweight'.")
```
